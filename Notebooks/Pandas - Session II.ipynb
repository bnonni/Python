### String Operations with Pandas

#### String operations.

```python
## Let's split the string to return the title
## n=1 specifies the number of times it splits the data based on the attribute ','

data['Title'] = data['Name'].str.split(',', n=1, expand=True)[1].str.split('.', expand=True)[0]
```
