```python
def a_inverse(a, modulus):
    # print(type(modulus))
    for i in range(modulus):
        if (((i * a) % modulus) == 1):
            return i
    return 0
```

```python
def GCD(inp1, inp2):
    while(inp1 != 0 and inp2 != 0):
        if(inp1 >= inp2):
            inp1 = inp1 - inp2
        else:
            inp2 = inp2 - inp1
    if(inp1 == 1 and inp2 == 0):
        return True
    elif(inp2 == 1 and inp1 == 0):
        return True;
    else: 
        return False
```

```python
def KeyGenerator(L4PN):
    a = 0
    b = 0
    a_inv = 0
    A_list = []
    for i in range(L4PN):
        if(GCD(i, L4PN)):
            A_list.append(i)
    rand_index = int(random() * len(A_list))
    a = A_list[rand_index]
    a_inv = a_inverse(A_list[rand_index], L4PN)
    b = rand_index
    k = str(a) + " " + str(a_inv) + " " + str(b)
    return k
```

```python
def encryption(key, content):
    key = key.split(" ")
    a = key[0]
    b = key[2]
#     print(a, b)
    cont = content.rstrip(' ')
    ASCcode = list(cont)
#     print(ASCcode)
    for i in range(len(ASCcode)):
        # print(ord(ASCcode[i]))
        # exit()
        temp = ((ord(ASCcode[i])) * a) + b
        # print(temp)
        ciphertext = temp + " "
    return ciphertext
```
