```python
#!/usr/bin/env python3
from random import *
```

```python
def a_inverse(a, modulus):
    for i in range(modulus+1):
        if i * (a % modulus) == 1:
            return i
        else:
            return 0
```

```python
def GCD(inp1, inp2):
    while(inp1 != 0 and inp2 != 0):
        if(inp1 >= inp2):
            inp1 = inp1 - inp2
        else:
            inp2 = inp2 - inp1
    if(inp1 == 1 and inp2 == 0):
        return True
    elif(inp2 == 1 and inp1 ==0):
        return True;
    else: return False
```

```python
def KeyGenerator(L4PN):
    a = 0
    b = 0
    a_inv = 0
    A_list = []
    for i in range(L4PN+1):
        if(GCD(i, L4PN)):
            A_list.append(i)
    rand_int = int(random() * len(A_list))
    a_inv = a_inverse(A_list[rand_int], L4PN)
    b =  random() * rand_int
    temp = str(a) + str(b) + str(a_inv)
    return temp
```
