__1. Maximum Posteriori Hypothesis (10 points)__

When a test for steroids is given to soccer players, 98% of the players taking steroids test positive and 12% of the players not taking steroids test positive. Suppose that 5% of soccer players take steroids. What is the maximum posteriori hypothesis for a soccer player who tests positive? What are the exact posterior probabilities?

__Max Posteriori Formula ($h_{MAP}$)__

$h_{MAP} = arg\,max_{h_i \in H} P(D|h)P(h)$

$h = steroids | \neg steroids$

$D = + | -$

<hr>

__Given:__ $P(steroids) = 0.05, \, P(\neg steroids) = 0.95$

__Q: What are the exact posterior probabilities?__

$P(+|\neg steroids) = 0.12 \; (false\;positive), \; P(-|\neg steroids) = 0.88 \; (true\;positive)$

$P(+|steroids) = 0.98 \; (true\,positive), \, P(-|steroids) = 0.02 \; (false\,negative)$

__Q: What is the maximum posteriori hypothesis for a soccer player who tests positive?__

$P(+|steroids)*P(steroids) = (0.98)(0.05) = 0.049 \approx 5\%$

$P(+|\neg steroids)*P(\neg steroids) = (0.12)(0.95) = 0.114 \approx 11\%$

$h_{MAP} = arg\,max_{h_i \in H} \{5\%, 11\%\}$

$\implies 11\% > 5\%$

$\therefore \, h_{MAP} = \neg steroids$

__Entropy (One Attribute) [$E(S)$]__

$E(S) = \sum_{i=1}^c - \,p_i log_2 p_i$

__Entropy (Multi Attributes) [$E(T,X)$]__

$E(T,X) = \sum_{c \in X} P(c) E(c) = P(X_1) \cdot c_1 log_2 c_1 + P(X_2) \cdot c_2 log_2 c_2 + ... + P(X_n) \cdot c_n log_2 c_n$

__Information Gain [$G(T,X)$]__

$G(T,X) = E(T) - E(T,X)$

__Best Split Point [$SplitInfo_A(D)$]__

$SplitInfo_A(D) = \sum_{j=1}^v \frac{|D_j|}{|D|} \cdot log_2 (\frac{|D_j|}{|D|})$

__Gain Ratio [$GainRatio(A)$]__

$GainRatio(A) = \frac{Gain(A)}{SplitInfo(A)}$

<hr>

__ID3 Algorithm__

__Play, Outlook__

$E(Play, Outlook) = P(Outlook) \cdot E(Outlook_{i_{p,\neg p}})$

$ = P(Sunny) \cdot E(Sunny_p, Sunny_{\neg p}) + P(Overcast) \cdot E(Overcast_p, Overcast_{\neg p}) + P(Rain) \cdot E(Rain_p, Rain_{\neg p})$

$ = (0.4)(1) + (0.1)(0.918) + (.5)(0.971) = 0.9773$

$G(Play,Outlook) = E(Play) - E(Play,Outlook) = E(5,5) - 0.9773 = 1 - 0.9773 = 0.0227 \approx 0.02$

__Play, Temperature__

$E(Play, Temperature) = P(Temperature) \cdot E(Temperature_{i_{p,\neg p}})$

$ = P(Hot) \cdot E(Hot_p, Hot_{\neg p}) + P(Mild) \cdot E(Mild_p, Mild_{\neg p}) + P(Cool) \cdot E(Cool_p, Cool_{\neg p})$

$ = (0.3)(0.918) + (0.3)(0.918) + (.4)(1) = 0.9508$

$G(Play,Temperature) = E(Play) - E(Play, Temperature) = E(5,5) - 0.9508 = 1 - 0.9508 = 0.0492 \approx 0.05$


__Play, Humidity__

$E(Play, Humidity) = P(Humidity) \cdot E(Humidity_{i_{p,\neg p}})$

$ = P(High) \cdot E(High_p, High_{\neg p}) + P(Normal) \cdot E(Normal_p, Normal_{\neg p})$

$ = (0.7)(0.985) + (0.3)(0.918) = 0.9649$

$G(Play,Humidity) = E(Play) - E(Play, Humidity) = E(5,5) - 0.9649 = 1 - 0.9649 = 0.0351 \approx 0.04 $


__Play, Wind__

$E(Play, Wind) = P(Wind) \cdot E(Wind_{i_{p,\neg p}})$

$ = P(Strong) \cdot E(Strong_p, Strong_{\neg p}) + P(Weak) \cdot E(Weak_p, Weak_{\neg p})$

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import export_graphviz
from sklearn.externals.six import StringIO
from IPython.display import Image
import pydotplus
```

```python
play_df = pd.DataFrame(pd.read_csv('data/Table3/table3.csv'))
play_df = play_df.drop('D', axis=1)
```

```python
play_df.head()
```

```python
play_df['Play_Label'] = play_df['Play'].apply(make_label)
play_df.head()
```

```python
le = LabelEncoder()
play_df['Outlook_Label'] = le.fit_transform(play_df.Outlook)
play_df['Temperature_Label'] = le.fit_transform(play_df.Temperature)
play_df['Humidity_Label'] = le.fit_transform(play_df.Humidity)
play_df['Wind_Label'] = le.fit_transform(play_df.Wind)
play_df.head(10)
```

```python
en_play_df = play_df[['Outlook_Label', 'Temperature_Label', 'Humidity_Label', 'Wind_Label', 'Play_Label']]
en_play_df.head(10)
```

```python
X = en_play_df.drop('Play_Label', axis=1)
y = en_play_df['Play_Label'].values
```

# Decision Tree
## Train_Test_Split = 67/33 Train/Test

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=0)
dtc = DecisionTreeClassifier()
dtc = dtc.fit(X_train, y_train)

print('Train Accuracy Score:', dtc.score(X_train, y_train))
print('Test Accuracy Score:', dtc.score(X_test, y_test))
```

```python
y_pred = dtc.predict(X_test)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
pd.crosstab(y_test, y_pred, rownames=['Real'], colnames=['Predicted'], margins=True)
```

```python
print(classification_report(y_test, y_pred))
```

```python
y_pred_proba = dtc.predict_proba(X_test)[:,1]
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
plt.plot([0,1],[0,1], 'k--')
plt.plot(fpr, tpr, label='KNN')
plt.xlabel('fpr')
plt.ylabel('tpr')
plt.title('DT ROC Curve')
plt.show()
```

```python
roc_auc_score(y_test, y_pred_proba)
```

```python
features = ['Outlook', 'Temperature', 'Humidity', 'Wind']
dot_data = StringIO()
export_graphviz(dtc, out_file=dot_data, filled=True, rounded=True, special_characters=True, feature_names=features, class_names=['No','Yes'])
```

```python
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
graph.write_png('dt_images/play_67-33.png')
Image(graph.create_png())
```
