```python
import math

def get_important_features(transformed_features, components_, columns):
    """
    This function will return the most "important" 
    features so we can determine which have the most
    effect on multi-dimensional scaling
    """
    num_columns = len(columns)

    # Scale the principal components by the max value in
    # the transformed set belonging to that component
    xvector = components_[0] * max(transformed_features[:,0])
    yvector = components_[1] * max(transformed_features[:,1])

    # Sort each column by it's length. These are your *original*
    # columns, not the principal components.
    important_features = { columns[i] : math.sqrt(xvector[i]**2 + yvector[i]**2) for i in range(num_columns) }
    important_features = sorted(zip(important_features.values(), important_features.keys()), reverse=True)
    return important_features

top_features = get_important_features(pca_xtrain1, pca.components_, x_train1.columns.values)
top20_features = top_features[:20]
print(top20_features)
print(top20_features[0][1])
```

```python
import math

def get_important_features(transformed_features, components_, columns):
    """
    This function will return the most "important" 
    features so we can determine which have the most
    effect on multi-dimensional scaling
    """
    num_columns = len(columns)

    # Scale the principal components by the max value in
    # the transformed set belonging to that component
    xvector = components_[0] * max(transformed_features[:,0])
    yvector = components_[1] * max(transformed_features[:,1])

    # Sort each column by it's length. These are your *original*
    # columns, not the principal components.
    important_features = { columns[i] : math.sqrt(xvector[i]**2 + yvector[i]**2) for i in range(num_columns) }
    important_features = sorted(zip(important_features.values(), important_features.keys()), reverse=True)
    return important_features

top_features = get_important_features(pca_xtest1, pca.components_, x_test1.columns.values)
top20_features = top_features[:20]
print(top20_features)
print(top20_features[0][1])
```

```python
# Create new data frame from top 20 features in x_test1
x_test1_20 = pd.DataFrame()

for i in range(20):
    x_test1_20[top20_features[i][1]] = x_test1[top20_features[i][1]]
    
x_test1_20
```

```python
# Columns don't match from x_train1_20 and x_test1_20. Thus, the columns that don't exist in both are biased
# to the sets making them irrelevant to the model

final_columns = []

for item in x_train1_20.columns.to_list():
    if item in x_test1_20.columns.to_list():
        final_columns.append(item)

final_columns = sorted(final_columns)
final_columns
```

```python
# Create new data frames from features in x_train1 and x_test1 that match as most significant
x_train1_final = pd.DataFrame()
x_test1_final = pd.DataFrame()

for i in range(len(final_columns)):
    x_train1_final[final_columns[i]] = x_train1[final_columns[i]]
    x_test1_final[final_columns[i]] = x_test1[final_columns[i]]
    
print(x_train1_final)
print(x_test1_final)
```

```python
# KNN to obtain y_test1 values
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors =5)

knn.fit(x_train1_final, y_train1)
y_test1_predicted = knn.predict(x_test1_final)
print(y_test1_predicted)
```

```python
# Create data frame for y_test1
y_test1 = pd.DataFrame({0: y_test1_predicted})
```

```python
knn.score(x_test1_final, y_test1_predicted)
```

```python
##### Logistic Regression #####
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(random_state=0)
lr.fit(x_train1_final, y_train1) 

lr_y_predict = lr.predict(x_test1_final)
lr_y_predict
```

```python
from sklearn import metrics
lr_cm = metrics.confusion_matrix(y_test1_predicted, lr_y_predict)
lr_report = metrics.classification_report(y_test1_predicted, lr_y_predict)
print(lr_cm)
print(lr_report)
```
