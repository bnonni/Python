### In-built Checks
#### 1. isnumeric - checks if whole string is numeric

```python
print('12345'.isnumeric())
print('12345.2'.isnumeric())
```

#### 2. isalpha

```python
print('21JumpStreet'.isalpha())
print('TwentyOneJumpStreet'.isalpha())
```

#### 3. isdigit

```python
print('123452'.isdigit())
print('LOWER'.isdigit())
```

#### 4. islower

```python
print('testing'.islower())
print('Testing'.islower())
```

```python
print('bry'.isupper())
print('BRY'.isupper())
```

#### 6. alnum

```python
print('testing123'.isalnum())
print('testing123!!!'.isalnum())
```

#### 7. isspace

```python
print(''.isspace())
print(' '.isspace())
print(' somethingelse'.isspace())
```

```python
thing = ' somethingelse'
thing[0]
count = 0
for char in thing:
    if thing[count].isspace():
        print("True")
    else:
        print("False")
    count = count+1
```
