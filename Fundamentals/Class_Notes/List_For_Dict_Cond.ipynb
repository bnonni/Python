# Loops

### For Loops

```python
letters = "ABCDEFGHIJKLMNOP"

for letter in letters:
    print(letter)
```

```python
for x in range(2,9):
    print(x)
```

```python
for tyuttt in range (2,9):
    print(tyuttt)
```

```python
for tyuttt in range (2,9):
    print("Bryan")
```

```python
numbers = [1,2,4,5,2,4,6,2,6]

for x in numbers:
    if x == 2:
        print(2)
```

```python
for index,value in enumerate(numbers):
    print(index)
    print(value)
```

```python
for index,value in enumerate(numbers):
    if index % 2 == 0:
        print(numbers[index])
    else:
        print("This does not fall within an even index")
```
